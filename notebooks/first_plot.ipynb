```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python

training_time_dict = {
    "ATTN-SUM-GAN (proposed method)": {
        "small": 5*60 + 5,
        "base": 10 * 60 + 3,
    },
    "SUM-GAN-AAE (Apostolidis et al.)": {
        "small": 1*3600 + 60*23 + 57,
        "base": 2*3600 + 60*18 + 4
    },
}
training_time_df = pd.DataFrame(training_time_dict)
training_time_df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=300)
training_time_df.plot(kind='bar', rot=0, ax=ax, colormap=matplotlib.cm.get_cmap('rainbow_r'))
ax.legend().set_visible(False)
ax.grid(axis='y')
ax.set_ylabel('Training time (s)')
# ax.set_xlabel('Model Configuration')

fig.legend(
    *ax.get_legend_handles_labels(), 
    loc='lower center', 
    ncol=2, 
    bbox_to_anchor=(0.5, 0),
    # frameon=False,
)

# plt.tight_layout()
plt.savefig("figs/training_time_plot.pdf")
```

```python
f1_score_dict = {
    "ATTN-SUM-GAN (proposed method)": {
        "small": 63.62,
        "base": 64.63
    },
    "SUM-GAN-AAE (Apostolidis et al.)": {
        "small": 62.91,
        "base": 62.10
    },
}
f1_score_df = pd.DataFrame(f1_score_dict)
f1_score_df.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 5), dpi=300)

training_time_df.plot(kind='bar', rot=0, ax=ax[0], colormap='rainbow')
ax[0].set_ylabel("Training time (s)")
ax[0].set_xlabel("Model Configuration")
ax[0].set_title("Training time (lower is better)")
ax[0].legend().set_visible(False)

f1_score_df.plot(kind='bar', rot=0, ax=ax[1], colormap='rainbow')
ax[1].set_ylabel("F1 score")
ax[1].set_xlabel("Model Configuration")
ax[1].set_title("Performance on TVSum (higher is better)")
ax[1].set_ylim([60, 66])
ax[1].legend().set_visible(False)

fig.legend(
    *ax[0].get_legend_handles_labels(), 
    loc='lower center', 
    ncol=2, 
    bbox_to_anchor=(0.5, -0.1),
    frameon=False,
    )

plt.tight_layout()
plt.savefig("figs/comparison_first_fig.pdf")
plt.show()
```

```python
data_dict = {
    "Training time": {
        "ATTN-SUM-GAN (proposed method)": 5*60 + 5,
        "SUM-GAN-AAE (Apostolidis et al.)": 1*3600 + 60*23 + 57
    },
    "Performance on TVSum": {
        "ATTN-SUM-GAN (proposed method)": 63.62,
        "SUM-GAN-AAE (Apostolidis et al.)": 62.91
    },
}

data_df = pd.DataFrame(data_dict)
data_df.transpose()
data_df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=300)
data_df.plot(kind='bar', secondary_y='Performance on TVSum', rot=0, ax=ax, colormap='rainbow')
ax.legend().set_visible(False)

ax1, ax2 = plt.gcf().get_axes()
ax2.set_ylim([62, 66])

fig.legend(
    # *ax.get_legend_handles_labels(), 
    ['Training time', 'Performance'],
    loc='lower center', 
    ncol=2, 
    bbox_to_anchor=(0.5, -0.1),
    frameon=False,
    )

plt.tight_layout()
plt.savefig("figs/comparison_first_fig.pdf")
plt.show()

plt.show()
```
