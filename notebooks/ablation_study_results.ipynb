```python
import matplotlib.pyplot as plt
import numpy as np
```

## Hidden size

```python
# different hidden sizes
# x axis is log scale

plt.figure(figsize=(8, 6), dpi=300)
hidden_sizes = [32, 64, 128, 256, 512]
f1_scores = [61.36, 63.86, 61.68, 63.62, 61.15]

plt.plot(hidden_sizes, f1_scores, 'o-', color='blue', label='F1 score')
plt.xlabel('Hidden size')
plt.xticks(hidden_sizes)
plt.ylabel('F1 score')
# show y-grid
plt.grid(axis='y')
# show value on each point
for i in range(len(hidden_sizes)):
    plt.text(
        (hidden_sizes[i] + 8) if i < len(hidden_sizes) - 1 else (hidden_sizes[i] - 54), 
        f1_scores[i], 
        f1_scores[i])
plt.title('Ablation study: hidden size')
plt.savefig('figs/hidden_size.pdf')
plt.show()
```

## Number of heads

```python
# different hidden sizes
# x axis is log scale

plt.figure(figsize=(8, 6), dpi=300)
num_heads = [2, 4, 8, 16, 32]
f1_scores = [63.92, 63.62, 60.66, 63.01, 61.7]

plt.plot(num_heads, f1_scores, 'o-', color='red', label='F1 score')
plt.xlabel('#Heads')
plt.xticks(num_heads)
plt.ylabel('F1 score')
# show y-grid
plt.grid(axis='y')
# show value on each point
for i in range(len(num_heads)):
    plt.text(
        (num_heads[i] + 1) if i < len(num_heads) - 1 else (num_heads[i] - 3), 
        f1_scores[i], 
        f1_scores[i])
plt.title('Ablation study: number of heads')
plt.savefig('figs/num_head.pdf')
plt.show()
```
